```python
import math
import pandas as pd
import numpy as np

# __Bisection__

# Starting values
p_0 = -2
p_1 = 1

# Function at hand
def f(x):
    "function defined in homework"
    y = x + math.exp(x)
    return y

# List to store our values of p
bisection = []
bisection.append(p_0)

for i in range(100):
    # compute f(p_0), f(p) and p
    f0 = f(p_0)
    p  = p_0 + (p_1-p_0)/2
    fp = f(p)
    
    # appending p to our results list
    bisection.append(p)
    
    # checking our tolerance condition
    if fp == 0 or abs((p-p_0)/p_0) < 0.0001:
        # if tolerance satisfied, exit the loop
        break
    else:
        i = i+1
        # otherwise, check if p should replace either a or b
        if f0*fp > 0:
            p_0 = p
        else:
            p_1 = p
```

```python
#__Fixed Point Iteration__

# Starting values (same as Bisection)
p_0 = -2
p_1 = 1

# List to store our values of p
fpi = []
fpi.append(p_0)

# Function at hand 
def g(x):
    "This is equivalent to g(x) = x - f(x) = x - x - exp(x)"
    y = -math.exp(x)
    return y

for i in range(100):
    
    # Compute value of g(p)
    p = g(p_0)
    
    # Append result to our list
    fpi.append(p)
    
    # checking our tolerance condition
    if abs((p-p_0)/p_0) < 0.0001:
        # if satisfied, exit program
        break
    else:
        # if not, continue
        i = i+1
        p_0 = p
```

```python
#_Newton-Raphson_

# Starting values (same as Bisection)
p_0 = -2
p_1 = 1

# List to store our values of p
newton = []
newton.append(p_0)

# Function at hand (in addition to f(x))
def f_(x):
    "this is the derivative of f(x) = x + exp(x)"
    y = 1 + math.exp(x)
    return y

for i in range(100):
    
    # Compute p
    p = p_0 - f(p_0)/f_(p_0)
    
    # Append to our list of values
    newton.append(p)
    
    # Check tolerance condition
    if abs((p-p_0)/p_0) < 0.0001:
        # if satisfied, exit program
        break
    else:
        # if not, continue
        i = i+1
        p_0 = p
```

```python
#_False Position_

# Starting values (same as Bisection)
p_0 = -2
p_1 = 1

# List to store our values of p
falsepos = []
falsepos.append(p_0)

for i in range(100):
    
    # Compute p
    q_0 = f(p_0)
    q_1 = f(p_1)
    p = p_1 - q_1*(p_1-p_0)/(q_1-q_0)
    
    # Append our value of p
    falsepos.append(p)
    
    # Check tolerance condition
    if abs((p-p_0)/p_0) < 0.0001 or abs((p-p_1)/p_1) < 0.0001:
        # if satisfied, exit program
        break
    else:
        # if not, continue
        i = i+1
        q = f(p)
        
        # if q_1 on other side of root from p, recode values
        if q*q_1 < 0:
            p_0 = p_1
        
        # assign q_1 and p_1 to root values
        p_1 = p
```

```python
# Let's put all the results together
max_l = max(len(bisection), len(fpi), len(newton), len(falsepos))

def extend_list(l, n):
    l.extend([np.nan] * n)
    l = l[:n]
    return l

q1_tbl = {'Bisection': extend_list(bisection, max_l),
         'Fixed Point Iteration': extend_list(fpi, max_l),
         'Newton-Raphason': extend_list(newton, max_l),
         'False Positive': extend_list(falsepos, max_l)}

q1_df = pd.DataFrame(q1_tbl)
```

Assuming we start each method with the same initial guess(es) of $p_0=-2$ and $p_1=1$ with the tolerance condition $|\frac{p_n-p_{n-1}}{p_{n-1}}| < 0.0001$, we found that the following number of iterations were necessary:
* Bisection: 16

To find the roots of $f(x) = 2sin(\pi x) + x$, we use $g(x) = \frac{1}{2}*sin^{-1}(\frac{-x}{2})$.

Further, because $f$ has several roots, the function $sin^{-1}$ is only defined on the interval $[-1,1]$, and we are only interested in roots on the interval $[1,2]$, we further specify $g$ as $g(x) = 2 + \frac{1}{2}sin^{-1}(\frac{-x}{2})$. 

```python
#__Fixed Point Iteration__

# Starting value
start = 1

# List to store our values of p
q3_list = []
q3_list.append(start)

# Function at hand 
def q3_fun(x):
    "This is equivalent to g(x) = x - f(x) => f(x) = x - g(x)"
    #y = -2*math.sin(x*math.pi)
    y = 2 + math.asin(-x/2)/math.pi
#     y = -2*math.cos(-math.pi*x/2)
#     y = -2*math.sin(x*math.pi-8*math.pi)
    return y

def do_fpi(val, func, list_to_append):
    """function to conduct fixed point iteration
    val -- initial starting point
    func -- function at hand
    list_to_append -- list to store root values
    """
    for i in range(100):

        # Compute value of g(p)
        root = func(val)

        # Append result to our list
        list_to_append.append(root)

        # checking our tolerance condition
        if abs(root-val) < 0.01:
            # if satisfied, exit program
            break
        else:
            # if not, continue
            if i == 99: print('convergence not achieved')
            i = i+1
            val = root
        
do_fpi(start, q3_fun, q3_list)

# return values
q3_df = pd.DataFrame({'Fixed Point Iteration':q3_list})
q3_df
```

We replace the ordinary annuity equation $A = \frac{P}{i} [1-(1+i)^{-n}]$ with the appropriate numbers such that $A = 135,000$, $P = 12,000$ (1,000 x 12 months) and $n = 30$.

As a result, we get: $135,000 = \frac{12,000}{i} [1-(1+i)^{-30}]$.

We define $f$ and $f'$ such that:

$f(x) = 135 - \frac{12}{x} + \frac{12}{x}(1+x)^{-30}$,

$f'(x) = \frac{12}{x^2} - \frac{360(1+x)^{-31}}{x} - \frac{12(1+x)^{-30}}{x^2}$

Solving this using the Newton-Raphson method, we get that $i \approx 0.0801$, in other words, the maximal interest rate is $8\%$.
